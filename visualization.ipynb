### Why choose Matplotlib and Seaborn ?
Both libraries are among the popular choices for visualization in python . heres why :
- Matplotlib : known for its flexibility , matplotlib allows detailed customization of visual elements . making it perfect fro tailored plots .
- Seaborn : Built on top of matplotlib , seaborn simplifies visualization tasks by offering high level functions to create aesthetically pleasing stat graphs .

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

```

```python
# line plot

x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.plot(x, y, label = "sine wave")
plt.title("Sine wave line plot")
plt.xlabel("value of x")
plt.ylabel("sine(x)")
plt.legend()
plt.show()
```

```python
#bar chart

categories = ['A', 'B', 'C']
values = [50, 100, 75]
plt.bar(categories, values, color=["red", "green", "blue"])
plt.title("Bar chart example")
plt.xlabel('categories')
plt.ylabel('values')
plt.show()

```

```python
#histogram
categories = ['A', 'B', 'C']
values = np.random.normal(170, 10, 250)
plt.hist(  values , color = ["red"])
plt.title("Bar chart example")
plt.xlabel('categories')
plt.ylabel('values')
plt.show()

```

```python
# histogram and kDE plot

data = np.random.randn((1000))
sns.histplot(data, kde=True, color='green')
plt.title("Histogram with KDE")
plt.show()
```

```python
# heatmap

matrix = np.random.rand(5, 5)
sns.heatmap(matrix, annot = True, cmap = "viridis")
plt.title('heatmap example')
plt.show()
```

```python
# pair plot

sns.set_style("whitegrid")
df = sns.load_dataset("iris")
sns.pairplot(df, hue = 'species', markers = ['o', 's', 'D'])
plt.show()
```
