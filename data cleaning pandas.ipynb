### Data Cleaning

Bad data can be
- empty data
- wrong data
- duplicates
- data in wrong format


```python
# clean pandas_dataset.csv

import pandas as pd
df = pd.read_csv("/content/drive/MyDrive/ml_from_scratch/data/pandas_dataset.csv")
df
```

```python
# lets see the shape of dataset
# total rows and columns

df.shape
```

```python
# lets examine the datatypes

data_types = df.dtypes
data_types
```

```python
# identify null values variables

isnull = df.isnull()
isnull
```

```python
isnull_count = isnull.sum()
print("total number of null values: ", isnull_count)
```

```python
# total nulls in dataset

total_isnull_count = isnull.sum().sum()
print("Total nulls : ", total_isnull_count)
```

```python
# replace single quote with blank and make the data formatted

df["Date"] = df["Date"].replace("'", "" , regex = True)
df
```

```python
# handle null values , return dataframe with no empty cells
# we can use df.dropna(), it simplyh removes empty cells but willnot change the dataframe,
# if we want dataframe to be changed , simply use inplace=True

new_df = df.dropna()
new_df
```

```python
isnull_count_newdf = new_df.isnull().sum().sum()
isnull_count_newdf
```

```python
# another way of handling null values is by replacing the null values

df.fillna(130, inplace=True)
df
```
