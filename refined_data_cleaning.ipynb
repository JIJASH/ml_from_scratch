##Eliminating the Null values

```python
# eliminating the null values

df[[ "Phone_Number", "Do_Not_Contact"]] = df[[ "Phone_Number", "Do_Not_Contact"]].fillna("N/a")
df["Last_Name"] = df["Last_Name"].fillna("unknown")
```

```python
df
```

##Eliminating the redundant and duplicates

```python
# eliminating the redundant values

duplicates = df[df.duplicated(keep = False)]
duplicates
```

```python
df = df.drop_duplicates()
df
```

##Eliminating non word characters from column : Last_Name

```python
df["Last_Name"] = df["Last_Name"].replace(r'[^a-zA-Z]' , '' , regex = True)
df
```

##Eliminating non-digit characters from the phone_number

```python
df["Phone_Number"] = df["Phone_Number"].replace(r'\D', '', regex = True )
df
```

```python
df["Phone_Number"] = df["Phone_Number"].replace("", "N/a", regex = True )
df
```

##Replacing Y and N with Yes and No in Paying Customer and Do_not_contact columns

```python
df["Paying Customer"] = df["Paying Customer"].replace({"Y" : "Yes", "N" : "No"})
df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace({"Y" : "Yes", "N" : "No"})
df
```
