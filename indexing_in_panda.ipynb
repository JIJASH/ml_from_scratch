## Group by and aggregating

```python
df_flavor = pd.read_csv("/content/drive/MyDrive/ml_from_scratch/data/Flavors.csv")
df_flavor
```

```python
group_by_frame = df_flavor.groupby("Base Flavor")
```

```python
type(group_by_frame)
```

```python
group_by_frame.mean(numeric_only = True)
```

```python
group_by_frame.count()
```

```python
group_by_frame.sum()
```

## Aggregating


```python
des = df_flavor.groupby(['Base Flavor', 'Liked']).agg({'Flavor Rating' : ['mean', 'max', 'count', 'sum'] , 'Texture Rating' : ['mean', 'max', 'count', 'sum']})
des
```

```python
des.describe()
```

## Data Merge ,concatenate and join

```python
import pandas as pd
```

```python
df1 = pd.read_csv("/content/drive/MyDrive/ml_from_scratch/data/LOTR.csv")
df1
```

```python
df2 = pd.read_csv("/content/drive/MyDrive/ml_from_scratch/data/LOTR 2.csv")
df2
```

```python
df1.merge(df2, how = 'inner', on = ['FellowshipID', 'FirstName'])
```

```python
df1.merge(df2, how = 'outer')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
# concatenate df1 and df2

pd.concat([df1,df2])
```
